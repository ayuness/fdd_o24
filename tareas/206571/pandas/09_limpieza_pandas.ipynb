Crea codigo para **guardar** y **cargar** un archivo JSON en el que se guarde la `estrategia` y `valor` que utilizaste para **imputar**. Por ejemplo: Si hay una columna que se llama `columna 3` y utilizaste la estrategia de imputacion de media, y existe otra llamada `columna 4` y  elegiste la palabra 'missing' el JSON debera contener:

 De tal manera que para cada columna que tenga un metodo de imputacion apunte a otro diccionario donde el **key** `estrategia` describa de manera sencilla el metodo, y el **key** `valor` el valor usado. En general:   
 `{'nombre de la columna':{'estrategia':'descripcion de estrategia', 'valor':'valor utilizado'}}`. 
 

De utilizar mas de un metodo puedes anidarlos en una lista  
  `[{...},{...}]`.  

Incluso si la columna utilizada no sufrio imputacion, es necesario que la agregues al JSON.

### Imputacion

```python
# Tu codigo aqui
imputation_strategies = {}

financial_cols = [
    'loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'installment', 'annual_inc',
    'revol_bal', 'total_pymnt', 'total_rec_prncp', 'total_rec_int', 'recoveries',
    'collection_recovery_fee', 'last_pymnt_amnt', 'total_rev_hi_lim'
]
for col in financial_cols:
    median_value = loans_copy[col].median()
    imputation_strategies[col] = {
        'estrategia': 'mediana',
        'valor': float(median_value) if pd.notnull(median_value) else None
    }

score_cols = [
    'fico_range_low', 'fico_range_high', 'dti', 'revol_util', 'bc_util',
    'il_util', 'all_util', 'pct_tl_nvr_dlq', 'percent_bc_gt_75'
]
for col in score_cols:
    mean_value = loans_copy[col].mean()
    imputation_strategies[col] = {
        'estrategia': 'media',
        'valor': float(mean_value) if pd.notnull(mean_value) else None
    }

count_cols = [
    'delinq_2yrs', 'inq_last_6mths', 'open_acc', 'pub_rec', 'total_acc',
    'collections_12_mths_ex_med', 'acc_now_delinq', 'chargeoff_within_12_mths',
    'num_accts_ever_120_pd', 'num_tl_30dpd', 'tax_liens'
]
for col in count_cols:
    imputation_strategies[col] = {
        'estrategia': 'valor_constante',
        'valor': 0
    }

time_cols = [
    'mths_since_last_delinq', 'mths_since_last_record', 'mths_since_last_major_derog',
    'mo_sin_old_il_acct', 'mths_since_recent_bc', 'mths_since_recent_inq'
]
for col in time_cols:
    max_val = loans_copy[col].max()
    imputation_strategies[col] = {
        'estrategia': 'maximo_plus_one',
        'valor': float(max_val + 1) if pd.notnull(max_val) else 0
    }

categorical_ordered_cols = ['grade', 'sub_grade']
for col in categorical_ordered_cols:
    mode_value = loans_copy[col].mode()[0]
    imputation_strategies[col] = {
        'estrategia': 'moda',
        'valor': str(mode_value) if pd.notnull(mode_value) else None
    }

categorical_cols = [
    'emp_title', 'home_ownership', 'verification_status', 'pymnt_plan',
    'purpose', 'title', 'application_type', 'hardship_type', 'hardship_reason'
]
for col in categorical_cols:
    imputation_strategies[col] = {
        'estrategia': 'valor_constante',
        'valor': 'Unknown'
    }

text_cols = ['desc']
for col in text_cols:
    imputation_strategies[col] = {
        'estrategia': 'valor_constante',
        'valor': ''
    }

date_cols = [
    'issue_d', 'earliest_cr_line', 'last_pymnt_d', 'next_pymnt_d',
    'last_credit_pull_d', 'hardship_start_date', 'hardship_end_date'
]
for col in date_cols:
    median_date = loans_copy[col].median()
    imputation_strategies[col] = {
        'estrategia': 'mediana_fecha',
        'valor': str(median_date) if pd.notnull(median_date) else None
    }

flag_cols = ['hardship_flag', 'debt_settlement_flag']
for col in flag_cols:
    imputation_strategies[col] = {
        'estrategia': 'valor_constante',
        'valor': 'N'
    }

location_cols = ['zip_code', 'addr_state']
for col in location_cols:
    imputation_strategies[col] = {
        'estrategia': 'valor_constante',
        'valor': 'Unknown'
    }

loans_copy
```

### Codigo para salvar y cargar JSONs

```python
# guardado 
import json
with open('estrategias.json', 'w') as f:
    json.dump(imputation_strategies, f, indent=4)

# carga
with open('estrategias.json', 'r') as f:
    data = json.load(f)


def aplicar_estrategias(df, estrategias):
    df_copy = df.copy()
    for columna, info in estrategias.items():
        if columna in df_copy.columns:
            if info['estrategia'] == 'mediana':
                df_copy[columna] = df_copy[columna].fillna(info['valor'])
            elif info['estrategia'] == 'media':
                df_copy[columna] = df_copy[columna].fillna(info['valor'])
            elif info['estrategia'] == 'valor_constante':
                df_copy[columna] = df_copy[columna].fillna(info['valor'])
            elif info['estrategia'] == 'maximo_plus_one':
                df_copy[columna] = df_copy[columna].fillna(info['valor'])
            elif info['estrategia'] == 'moda':
                df_copy[columna] = df_copy[columna].fillna(info['valor'])
            elif info['estrategia'] == 'mediana_fecha':
                df_copy[columna] = pd.to_datetime(df_copy[columna], errors='coerce')
                df_copy[columna] = df_copy[columna].fillna(pd.to_datetime(info['valor']))
    return df_copy
```
